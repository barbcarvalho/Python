```python

def validar_colog(codigo):
    if not codigo.isdigit() or len(codigo) != 15:
        print("Colog está inválido! Revise o que foi digitado, o Colog deve conter 15 números.")
        return

    base_num = []
    inicio = codigo[:13]
    for i in inicio:
        dig = int(i)
        base_num.append(dig)


    dv_inf_1 = int(codigo[13])
    dv_inf_2 = int(codigo[14])

    peso_dv1 = list(range(2, 15))
    soma_dv1 = 0
    j = 0
    for i in range(len(base_num) - 1, -1, -1):
        digito = base_num[i]
        peso = peso_dv1[j]
        produto = digito * peso
        soma_dv1 = soma_dv1 + produto
        j = j + 1

    resto_dv1 = soma_dv1 % 11
    if resto_dv1 < 2:
        dv1 = 0
    else:
        dv1 = 11 - resto_dv1

    base_com_dv1 = []
    for b in base_num:
        base_com_dv1.append(b)
    base_com_dv1.append(dv1)

    pesos_dv2 = list(range(3, 17))
    soma_dv2 = 0
    j = 0
    for i in range(len(base_com_dv1) - 1, -1, -1):
        digito = base_com_dv1[i]
        peso = pesos_dv2[j]
        produto = digito * peso
        soma_dv2 = soma_dv2 + produto
        j = j + 1

    resto_dv2 = soma_dv2 % 11
    if resto_dv2 < 2:
        dv2 = 0
    else:
        dv2 = 11 - resto_dv2

    if dv1 == dv_inf_1 and dv2 == dv_inf_2:
        print("Colog válido!!")
    else:
        print("Colog inválido!")


codigo_digitado = input("Digite o código Colog (15 números): ")
validar_colog(codigo_digitado)
```

```python
continuar = 'S'
while continuar =='S':
    nome = input("Digite o nome do cliente: ")
    lista = []
    quantidade = int(input("Digite a quantidade de produtos comprados: "))
    for i in range(1,(quantidade+1)):
        valor=float(input(f"Qual o valor do produto {i}?"))
        lista.append(valor)
    valor_total = sum(lista)
    pagamento = int(input("Forma de pagamento (1 - Dinheiro | 2 - Cartão | 3 - Pix) : "))
    if pagamento == 1:
        valor_nota = valor_total * 0.9
        forma = 'Dinheiro'
    elif pagamento == 2:
        valor_nota = valor_total
        forma = 'Cartão'
    elif pagamento == 3:
        valor_nota = valor_total * 0.95
        forma = 'Pix'
    print ("--- NOTA FISCAL ---")
    print (f'Cliente: {nome}')
    print (f'Total sem desconto : {valor_total}')
    print (f'Forma de pagamento : {forma}')
    print (f'Total a pagar : {valor_nota:.2f}')
    print('-------------------')
    continuar=(input("Deseja continuar S/N? "))
```
